```python
import pandas as pd
```

```python
fert = pd.read_csv('./data/gapminder_total_fertility.csv', index_col=0)
life = pd.read_excel('./data/gapminder_lifeexpectancy.xlsx', index_col=0, nrows=260)
population = pd.read_excel('./data/gapminder_population.xlsx', index_col=0)
```

```python
continents = pd.read_csv('./data/continents.csv', sep=";")
continents
```

```python
population.dropna(inplace=True)
fert.dropna(inplace=True)
life.dropna(inplace=True)
```

```python
fert.head()
```

```python
life.head()
```

```python
population.head()
```

```python
fert.columns = fert.columns.astype(int)
fert.columns
```

```python
life.columns = life.columns.astype(int)
life.columns
```

```python
fert.index.name = 'country'
life.index.name = 'country'
population.index.name = 'country'
```

```python
fert = fert.reset_index()
life = life.reset_index()
population = population.reset_index()
```

```python
fert = fert.melt(id_vars='country', var_name='year', value_name='fertility_rate')
```

```python
life = life.melt(id_vars='country', var_name='year', value_name='life_expectency')
```

```python
population = population.melt(id_vars='country', var_name='year', value_name='population')
```

```python
df = fert.merge(population)
```

```python
df = df.merge(life)
```

```python
df
```

```python
df['country'].nunique()
```

```python
df.shape
```

```python
continents.shape
```

```python
df = pd.merge(continents, df, how="inner", on="country")
df
```

```python
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(12,9)})

df_subset = df.loc[df['year'] == 2000]
sns.scatterplot(
    x='life_expectency', 
    y='fertility_rate',
    size='population',
    hue='continent',
    data=df_subset, 
    alpha=0.6
)
```

```python
sns.set(rc={'figure.figsize':(12,9)})
for year in range(1940, 2015):
    plt.axis((20, 90, 0, 10))
    df_subset = df.loc[df['year'] == year]
    sns.scatterplot(
        x='life_expectency', 
        y='fertility_rate', 
        size='population',
        hue='continent',
        data=df_subset, 
        alpha=0.6
    )
    plt.savefig(fname="/tmp/lifeexp_" + str(year) + ".png")
    plt.close()
```

```python
import imageio

images = []

for i in range(1960, 2015):
    filename = '/tmp/lifeexp_{}.png'.format(i)
    images.append(imageio.imread(filename))

imageio.mimsave('./output/animated-scatterplot.gif', images, fps=3)
```
