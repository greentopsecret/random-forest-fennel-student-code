```python
stocks_all = pd.read_csv('../data/stock_px.csv', parse_dates=True, index_col=0)
stocks = stocks_all[['AAPL', 'MSFT', 'XOM']]
stocks
```

```python
stocks.resample('W').first()
```

```python
stocks.resample('2W').mean()
```

```python
stocks.resample('M').sum()
```

```python
stocks.resample('6h').sum()
```

```python
stocks.resample('6h').ffill()
```

```python
stocks = stocks.asfreq('B').fillna(method='ffill')

stocks['AAPL'].plot()
```

```python
stocks['AAPL'].rolling(25).mean().plot()
```

```python
stocks['AAPL'].rolling(25, min_periods=10).std().plot()
```
